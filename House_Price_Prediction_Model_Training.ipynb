```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Loading the dataset**

```python
import urllib.request

# Step 1: Load the data from the URL
url = "https://lib.stat.cmu.edu/datasets/boston"
raw_data = urllib.request.urlopen(url).read().decode("utf-8")

# Step 2: Split the raw text into lines
lines = raw_data.split("\n")

# Step 3: Extract the actual data (starts from line 22)
data_lines = lines[22:]  # Skip header info
data_lines = [line for line in data_lines if line.strip() != '']  # Remove empty lines

# Step 4: Parse data lines (2 lines per data point)
data = []
for i in range(0, len(data_lines), 2):
    first_line = list(map(float, data_lines[i].strip().split()))
    second_line = list(map(float, data_lines[i + 1].strip().split()))
    data.append(first_line + second_line)

# Step 5: Convert to DataFrame
columns = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM",
    "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
]
df = pd.DataFrame(data, columns=columns)
```

```python
# Dividing into independent and dependent

X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
# Standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

**Training the Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,Y_train)
```

```python
# print the coefficient and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# on which parameters the model has been trained
regression.get_params()
```

```python
# prediction with the Test Data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
# plot a scatter plot for the prediction
plt.scatter(Y_test,reg_pred)
```

```python
residuals = Y_test - reg_pred
```

```python
residuals
```

```python
# Ploting this residuals
sns.displot(residuals,kind='kde')
```

```python
# Scatter plot with respect to the predictions and residuals
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(Y_test,reg_pred))
print(mean_squared_error(Y_test,reg_pred))
print(np.sqrt(mean_squared_error(Y_test,reg_pred)))
```

```python
# R sqaure and adjusted R square
from sklearn.metrics import r2_score
score=r2_score(Y_test,reg_pred)
print(score)
```
